# 4. Hadley Wickham's Tidy Data Principles

Let me introduce to you some principles. These principles come from [Hadley Wickham's Journal of Statistical Software article](https://vita.had.co.nz/papers/tidy-data.pdf). If you are new member of Data Science, I suggest to you should read this arictle.

Tidy data is a standard way of mapping the meaning of a dataset to its structure. A dataset is messy or tidy depending on how rows, columns and tables are matched up with observations, variables and types. In tidy data:

1. Each variable forms a column.
2. Each observation forms a row.
3. Each type of observational unit forms a tab

```r
# New Variables
fbi$Region <- NA

fbi$Geographic_Division <- NA

fbi <- fbi[c(14:15, 1:13)]

fbi <- fbi  %>% mutate(Area = str_trim(Area))

head(fbi)
```

```r
################################ Northeast ################################ 

fbi$Region[
  which(fbi$Area %in% c(
    "Connecticut", "Maine", "Massachusetts", "New Hampshire", "Rhode Island6", "Vermont", "New Jersey", "New York6", "Pennsylvania"
    )
      )
        ] <- "Northeast"

#### Northeast - New England
fbi$Geographic_Division[
  which(fbi$Area %in% c(
    "Connecticut", "Maine", "Massachusetts", "New Hampshire", "Rhode Island6", "Vermont"
  )
    )
      ] <- "New England"
                         
#### Northeast - Middle Atlantic                     
fbi$Geographic_Division[
  which(fbi$Area %in% c(
    "New Jersey", "New York6", "Pennsylvania"
  )
    )
      ] <- "Middle Atlantic"



################################ Midwest ################################ 

fbi$Region[
  which(fbi$Area %in% c(
    "Illinois", "Indiana6", "Michigan", "Ohio6", "Wisconsin6", "Iowa", "Kansas", "Minnesota", "Missouri", "Nebreska6", "North Dakota", "South Dakota6"
  )
    )
      ] <- "Midwest"

#### Midwest - East North Central
fbi$Geographic_Division[
  which(fbi$Area %in% c(
  "Illinois", "Indiana6", "Michigan", "Ohio6", "Wisconsin6"
  )
    )
      ] <- "East North Central"

#### Midwest - West North Central
fbi$Geographic_Division[
  which(fbi$Area %in% c(
  "Iowa", "Kansas", "Minnesota", "Missouri", "Nebreska6", "North Dakota", "South Dakota6"
  )
    )
      ] <- "West North Central"




################################ South ################################ 

fbi$Region[
  which(fbi$Area %in% c(
  "Delaware", "District of Columbia7", "Florida", "Georgia8", "Maryland", "North Carolina6", "South Carolina", "Virginia", "West Virginia","Alabama", "Kentucky", "Mississippi6", "Tennessee", "Arkansas", "Louisiana", "Oklahoma6", "Texas"
  )
    )
      ] <- "South"

#### South - South Atlantic 
fbi$Geographic_Division[
  which(fbi$Area %in% c(
  "Delaware", "District of Columbia7", "Florida", "Georgia8", "Maryland", "North Carolina6", "South Carolina", "Virginia", "West Virginia"
  )
    )
      ] <- "South Atlantic"

#### South - East South Central
fbi$Geographic_Division[
  which(fbi$Area %in% c(
  "Alabama", "Kentucky", "Mississippi6", "Tennessee"
  )
    )
      ] <- "East South Central"

#### South - West South Central
fbi$Geographic_Division[
  which(fbi$Area %in% c(
    "Arkansas", "Louisiana", "Oklahoma6", "Texas"
      )
        )
          ] <- "West South Central"



################################ West ################################ 

fbi$Region[
  which(fbi$Area %in% c(
    "Arizona", "Colorado", "Idaho", "Montana", "Nevada", "New Mexico6", "Utah", "Wyoming", "Alaska", "California", "Hawaii", "Oregon6", "Washington6", "Puerto Rico9"
    )
      )
         ] <- "West"

#### West - Mountain
fbi$Geographic_Division[
  which(fbi$Area %in% c(
    "Arizona", "Colorado", "Idaho", "Montana", "Nevada", "New Mexico6", "Utah", "Wyoming"
    )
     )
        ] <- "Mountain"

#### West - Pacific
fbi$Geographic_Division[
  which(fbi$Area %in% c(
    "Alaska", "California", "Hawaii", "Oregon6", "Washington6", "Puerto Rico9"
    )
      )
        ] <- "Pacific"



################################ TOTAL ################################ 

# ve Amerika toplamını da Total olarak gösterelim.
fbi$Region[which(fbi$Area == "United States Total6, 7, 8, 9")] <- "Total"

fbi$Geographic_Division[which(fbi$Area == "United States Total6, 7, 8, 9")] <- "Total"

head(fbi, 20)
```

## 7.6. Crime Statistics

We need crime statistics of the states. 

```r
# Total
fbi_total <- fbi %>% filter(Region == "Total")

# Region
fbi_region <- fbi %>% filter(Area %in% c("Northeast6", "Midwest6", "South6, 7 ,8", "West6"))

# Geographic Division
fbi_geo_div <- fbi %>% filter(Geographic_Division %in% c("New England","Middle Atlantic","East North Central","West North Central",
                                                         "South Atlantic","East South Central","West South Central","Mountain","Pacific") )

# State
fbi_state <- fbi %>% 
  filter(Region != "Total", !is.na(Region)) 

# Percent Change
fbi_percent_change <- fbi_state %>% 
  filter(Year %in% "Percent change")

# States Data
fbi_state <- fbi_state %>% 
  filter(Year != "Percent change",
         Area != c("New England6","Middle Atlantic6","East North Central6","West North Central6",
                   "South Atlantic6, 7, 8","East South Central6","West South Central6",
                   "Mountain6","Pacific6")
         )

head(fbi_state)
```

<a id="subsection8"></a>
## 7.7. String Manipulation

```r
for(i in c(",","6","7","8","9")){
    fbi_state$Area <- str_remove_all(fbi_state$Area, i)
}
```

<a id="subsection9"></a>
## 7.8. Transform Data Types

```r
fbi_state <- fbi_state %>% 
    mutate_at(vars(Population2:`Motor vehicle theft` ),funs(as.numeric)) %>% 
    rename(State = Area) %>%
    mutate_at(vars(c(State, Region:Year)), funs(as.factor))
```

# 9. Clustering

### 2015

```r
clustering <- fbi_state %>% 
    filter(Year == 2015) %>%
    select(-Year) %>%
    mutate_at(vars(c(Region, Geographic_Division)),funs(as.numeric)) %>%
    column_to_rownames("State")

# Apply Kmeans
km2 <- kmeans(clustering, 2)
km3 <- kmeans(clustering, 3)
km4 <- kmeans(clustering, 4)
km5 <- kmeans(clustering, 5)
km6 <- kmeans(clustering, 6)
km7 <- kmeans(clustering, 7)

# Visualization
grid.arrange(
    fviz_cluster(km2, clustering,main = "k=2"),
    fviz_cluster(km3, clustering,main = "k=3"),
    fviz_cluster(km4, clustering,main = "k=4"),
    fviz_cluster(km5, clustering,main = "k=5"),
    fviz_cluster(km6, clustering,main = "k=6"),
    fviz_cluster(km7, clustering,main = "k=7")
)
```

### 2016

```r
clustering <- fbi_state %>% 
    filter(Year == 2016) %>%
    select(-Year) %>%
    mutate_at(vars(c(Region, Geographic_Division)),funs(as.numeric)) %>%
    column_to_rownames("State")

# Apply Kmeans
km2 <- kmeans(clustering, 2)
km3 <- kmeans(clustering, 3)
km4 <- kmeans(clustering, 4)
km5 <- kmeans(clustering, 5)
km6 <- kmeans(clustering, 6)
km7 <- kmeans(clustering, 7)

# Visualization
grid.arrange(
    fviz_cluster(km2, clustering,main = "k=2"),
    fviz_cluster(km3, clustering,main = "k=3"),
    fviz_cluster(km4, clustering,main = "k=4"),
    fviz_cluster(km5, clustering,main = "k=5"),
    fviz_cluster(km6, clustering,main = "k=6"),
    fviz_cluster(km7, clustering,main = "k=7")
)
```
